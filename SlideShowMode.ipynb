<script>
var SlowShow = function(cell){
    this.cell = cell;
    this.cm = cell.code_mirror;
    this.state = 0;
}
    
SlowShow.prototype.nline =function(){
    return this.cell.code_mirror.lineCount()
}

SlowShow.prototype.hall = function(){
    //var cell = IPython.notebook.get_cell(0)
    //var nline = cell.code_mirror.lineCount()
    cm = this.cm;
    // bug in CM do not hide all lines
    for(var i = 0; i <this.nline()-1; i++)
    { 
        cm.hideLine(cm.getLineHandle(i))
    }
}

SlowShow.prototype.sall = function(){
    //var cell = IPython.notebook.get_cell(0)
    //var nline = cell.code_mirror.lineCount()
    cm = this.cm;
    // bug in CM do not hide all lines
    for(var i = 0; i <this.nline()-1; i++)
    { 
        cm.showLine(cm.getLineHandle(i))
    }
}

SlowShow.prototype.hmark = function(){
    var cell = this.cell;
    var cm = this.cm;
    for(var i = 0; i < this.nline() ; i++){ 
        if(cm.getLine(i).match(/ *##----/))
        {
            cm.hideLine(cm.getLineHandle(i))
        }
    }
}
    
SlowShow.prototype.pshow = function(start){
    var cell = this.cell;
    var cm = this.cm;
    for(var i = this.state ; i < this.nline() ; i++){ 
        if(cm.getLine(i).match(/ *##----/))
        {
            this.state = i+1;
            return
        } else {
            cm.showLine(cm.getLineHandle(i))
        }
       this.state = 0;
    }
}
    
var bnb = function(){
    ss = new SlowShow(IPython.notebook.get_cell(10))
    ss.hall()
    return false;
}
var fu = function(){ss.pshow(); return false;};
var rv = function(){ss.sall(); return false;};
</script>
